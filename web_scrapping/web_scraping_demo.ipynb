```python
#all links want to store inside a list
all_links=soup.find_all("a")
all_link_list=[]
for link in all_links:
    all_link_list.append(link.get("href"))
```

```python
#want to get only this wikitable sortable
right_table = soup.find('table',class_='wikitable sortable')
```

```python
#now i am going to see the right table data
right_table
```

```python
#grab all the links from this right table
right_table_link = right_table.find_all('a')
```

```python
right_table_link
```

```python
#create one list and add all the titles inside this list
country=[]
for links in right_table_link:
    country.append(links.get('title'))
```

```python
country
```

```python
print(len(country))
```

```python
#now we can create the area and country both
all_rows=right_table.findAll('tr')
all_rows
```

```python
list_area=[]
country=[]
#all_rows
#now we can loop through it
for row in all_rows:
    col=row.find_all('td')
    if len(col) > 2:
        if col[1].find('a') is not None:
           span_elements =col[3].text
           #print(span_elements)
           title=col[1].find('a').get('title')
           if title is not None:
              country.append(title)
              list_area.append(span_elements)
```
