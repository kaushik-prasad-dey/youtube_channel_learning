```python
#!ngrok http 80
#!ngrok.connect(5000)
from pyngrok import ngrok
# Connect to ngrok and create a public tunnel for port 5000
public_url = ngrok.connect(port=5000)
# Print the ngrok tunnel information
print(' * ngrok tunnel "{}" -> "http://127.0.0.1:{}/"'.format(public_url, 5000))
```

```python
from flask import Flask
from flask_ngrok import run_with_ngrok
app=Flask(__name__)
run_with_ngrok(app)

@app.route('/')
def hello_world():
    return "welcome to the Flask Environment"

app.run()
```
